### Getting started with NLP

This notebook shows the text analysis with mainly SpaCy and NLTK(only for the stemming)

Thisn Notebook mainly focuses on simple text analysis.
* Tokanization
* POS Tagging
* lemmatization
* Stemming
* Named entity recognition

```python
text = 'Thank you so much for being here, We surely gonna learn cool things today!!'
```

```python
import spacy

model = spacy.load('en_core_web_sm')

model_op = model(text)
```

### Tokanization

```python
import spacy

model = spacy.load('en_core_web_sm')

model_op = model(text)

for token in model_op:
    print(token.text)
```

### Stop Words

```python
import spacy

sp = spacy.load('en_core_web_sm')

all_stopwords = sp.Defaults.stop_words

print(all_stopwords)
```

### POS Tagging

```python
for token in model_op:
    print(token.text + ' : ' + token.pos_)
```

```python
for token in model_op:
    print(token.text + ' : ' + token.lemma_)  # Lemma for Spacy is so intellagent
```

```python
for token in model_op:
    print(token.text + ' : ' + token.shape_)
```

```python
from spacy import displacy


displacy.serve(model_op, style = 'dep')
```

### Named Entity

Spacy supports the following entity types

| Type      | Description                                         |
| :-------  | :-------------------------------------------------- |
| PERSON    | People, including fictional.                        |
| NORP      | Nationalities or religious or political groups.     |
| EVENT     | Named hurricanes, battles, wars, sports events, etc.|
| LANGUAGE  | Any named language.                                 |
| DATE      | Absolute or relative dates or periods.              |
| MONEY     | Monetary values, including unit.                    |


for more you can visit here
[Named Entity](https://spacy.io/api/annotation#named-entities)


```python
text_2 = 'European authorities fined Google a record $5.1 billion on Wednesday for abusing its power in the mobile phone market and ordered the company to alter its practices'

model_op_1 = model(text_2)

for token in model_op_1.ents:
    print(token.text + ' : ' + token.label_)
```

```python
displacy.serve(model_op_1, style = 'ent')
```

```python
text_3 = u'Great greater gretest happy happier happiest'

model_op_2 = model(text_3)

for token in model_op_2:
    print(token.text + ' : ' + token.lemma_)
```

### Stemming

```python
from nltk.stem import PorterStemmer 
from nltk.tokenize import word_tokenize 
   
ps = PorterStemmer() 
   
words = ["program", "programs", "programer", "programing", "programers"] 
  
for w in words: 
    print(w, " : ", ps.stem(w)) 
```
